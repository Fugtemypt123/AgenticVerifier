```python
import json
import os
from pptx import Presentation

base_dir = '/home/shaofengyin/AutoPresent/generate/examples'
names = ['art_photos', 'business', 'design', 'entrepreneur', 'environment', 'food', 'marketing', 'social_media', 'technology']
ref_eval_result = {'match': 0, 'text': 0, 'color': 0, 'position': 0}
ref_free_eval_result = {'text': 0, 'image': 0, 'layout': 0, 'color': 0}
fail_num = 0
total_num = 0

for name in names:
    slide_path = os.path.join(base_dir, name, name + '.pptx')
    prs = Presentation(slide_path)
    pages_num = len(prs.slides)
    print(f"{name} {pages_num}")
    for slide_num in range(1, pages_num + 1):
        current_base_dir = os.path.join(base_dir, name, f"slide_{slide_num}", "refined_gpt4o")
        files = os.listdir(current_base_dir)
        files = sorted(files, key=lambda x: int(x), reverse=True)
        for file in files:
            ref_eval = os.path.join(current_base_dir, file, "ref_eval.txt")
            ref_free_eval = os.path.join(current_base_dir, file, "refine_eval.json")
            if os.path.exists(ref_eval) and os.path.exists(ref_free_eval):
                break   
        if not os.path.exists(ref_eval) or not os.path.exists(ref_free_eval):
            print(f"Refined GPT-4o not exist, using original GPT-4o instead")
            ref_eval = os.path.join(base_dir, name, f"slide_{slide_num}", "ref_eval.txt")
            ref_free_eval = os.path.join(base_dir, name, f"slide_{slide_num}", "gpt_4o_eval.json")
        if os.path.exists(ref_eval) and os.path.exists(ref_free_eval):
            with open(ref_eval, 'r') as f:
                current_ref_eval_result = f.read()
                current_ref_eval_result = current_ref_eval_result.split('\n')
                for result in current_ref_eval_result:
                    if 'match' in result:
                        ref_eval_result['match'] += float(result.split(': ')[1])
                    elif 'text' in result:
                        ref_eval_result['text'] += float(result.split(': ')[1])
                    elif 'color' in result:
                        ref_eval_result['color'] += float(result.split(': ')[1])
                    elif 'position' in result:
                        ref_eval_result['position'] += float(result.split(': ')[1])
            with open(ref_free_eval, 'r') as f:
                current_ref_free_eval_result = json.load(f)
                ref_free_eval_result['text'] += current_ref_free_eval_result['text']['score'] * 20
                ref_free_eval_result['image'] += current_ref_free_eval_result['image']['score'] * 20
                ref_free_eval_result['layout'] += current_ref_free_eval_result['layout']['score'] * 20
                ref_free_eval_result['color'] += current_ref_free_eval_result['color']['score'] * 20   
            # print(f"{name} slide_{slide_num} {ref_eval_result} {ref_free_eval_result}")
        else:
            fail_num += 1
            # print(f"{name} slide_{slide_num} failed")
        total_num += 1
        
for key in ref_eval_result.keys():
    ref_eval_result[key] = ref_eval_result[key] / total_num
for key in ref_free_eval_result.keys():
    ref_free_eval_result[key] = ref_free_eval_result[key] / total_num

print(f"success_rate: {(total_num - fail_num) / total_num}")
print(f"ref_eval_result: {ref_eval_result}")
print(f"ref_free_eval_result: {ref_free_eval_result}")
print(f"Overall: {(ref_eval_result['match'] + ref_eval_result['text'] + ref_eval_result['color'] + ref_eval_result['position'] + ref_free_eval_result['text'] + ref_free_eval_result['image'] + ref_free_eval_result['layout'] + ref_free_eval_result['color']) / 8}")
```

```python
import os
import shutil

# put files in dir
base_path = "/home/shaofengyin/AutoPresent/generate/examples"
files = os.listdir(base_path)
for file in files:
    if os.path.isdir(os.path.join(base_path, file)):
        deck_path = os.path.join(base_path, file)
        files = os.listdir(deck_path)
        for file in files:
            if os.path.isdir(os.path.join(deck_path, file)):
                slide_path = os.path.join(deck_path, file)
                refined_name = os.path.join(slide_path, "baseline")
                os.makedirs(refined_name, exist_ok=True)
                eval_name = os.path.join(slide_path, "gpt_4o_eval.json")
                ref_eval = os.path.join(slide_path, "ref_ori_eval.txt")
                jpg = os.path.join(slide_path, "gpt_4o.jpg")
                md = os.path.join(slide_path, "gpt_4o.md")
                pptx = os.path.join(slide_path, "gpt_4o.pptx")
                py = os.path.join(slide_path, "gpt_4o.py")
                if os.path.exists(ref_eval):
                    shutil.move(ref_eval, os.path.join(refined_name, "ref_eval.txt"))
                if os.path.exists(eval_name):
                    shutil.move(eval_name, os.path.join(refined_name, "refree_eval.json"))
                if os.path.exists(jpg):
                    shutil.move(jpg, os.path.join(refined_name, "gpt_4o.jpg"))
                if os.path.exists(md):
                    shutil.move(md, os.path.join(refined_name, "gpt_4o.md"))
                if os.path.exists(pptx):
                    shutil.move(pptx, os.path.join(refined_name, "gpt_4o.pptx"))
                if os.path.exists(py):
                    shutil.move(py, os.path.join(refined_name, "gpt_4o.py"))
                    
```
